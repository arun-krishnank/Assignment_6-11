1. Create a pie chart presenting the male/female proportion

```python
%matplotlib inline
import pandas as pd
import matplotlib.pyplot as plt

titanic = pd.read_csv("https://raw.githubusercontent.com/Geoyi/Cleaning-Titanic-Data/master/titanic_original.csv")
male_count = titanic[titanic['sex'] == 'male']['sex'].count()
female_count = titanic[titanic['sex'] == 'female']['sex'].count()
total_count = titanic['sex'].count()

slices=[male_count,female_count]
gender_count = ['Proportion of Males','Proportion of Females']
cols=['c','m']

plt.title("Pie chart presenting the male/female proportion")
plt.pie(slices,labels=gender_count,colors=cols,autopct='%1.1f%%')
plt.show()
```

2. Create a scatterplot with the Fare paid and the Age, differ the plot color by gender

```python
%matplotlib inline
colors = ['green','blue']
gender = ['male','female']

for color,val in zip(colors,gender):
    plt.scatter(titanic[titanic['sex'] == val]['age'],titanic[titanic['sex'] == val]['fare'],c=color,label=val.title())

plt.title("Fare Paid vs Age")
plt.xlabel("Age")
plt.ylabel("Fare Paid")
plt.legend(loc='upper right')
plt.show()
```
