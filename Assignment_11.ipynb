# Problem statement 2:
Read the following data set:
https://archive.ics.uci.edu/ml/machine-learning-databases/adult/adult.data

```python
import pandas as pd
column_names=['age','workclass','fnlwgt','education','education-num','marital-status','occupation','relationship','race','sex','capital-gain','capital-loss','hours-per-week','native-country']
sqldf = pd.read_csv("https://archive.ics.uci.edu/ml/machine-learning-databases/adult/adult.data",header=None,names=column_names,index_col=False)
sqldf.head(10)
```

Task:
1. Create an sqlalchemy engine using a sample from the data set

```python
from sqlalchemy import create_engine, MetaData, Table, Column, Integer, String
from sqlalchemy.orm import sessionmaker
from sqlalchemy import MetaData

engine = create_engine('sqlite://')
meta = MetaData()
ad_data=Table(
    'adultdata', meta,
    Column('age', Integer),
    Column('workclass', String),
    Column('education', String),
    Column('maritalstatus',String),
    Column('occupation',String)
)
meta.create_all(engine)

conn = engine.connect()

for index,row in sqldf.head(10).iterrows():
    ins = ad_data.insert().values(age=row.age, workclass=row.workclass, education=row.education, maritalstatus=row['marital-status'],occupation=row.occupation)
    conn.execute(ins)
    
for row in conn.execute(ad_data.select()):
    print(row)
```

2. Write two basic update queries

```python
upd = ad_data.update().where(ad_data.c.maritalstatus == ' Never-married').values(maritalstatus='Bachelor')
conn.execute(upd) # Updating marital-status for Never-married to Bachelor
upd2 = ad_data.update().where(ad_data.c.workclass != ' Private').values(workclass='Public')
conn.execute(upd2) # Updating all workclass which are not Private as Public

for row in conn.execute(ad_data.select()):
    print(row)
```


3. Write two delete queries

```python
del1 = ad_data.delete().where(ad_data.c.age > 50)
conn.execute(del1)
s = ad_data.select()
print("Result after deleting records of adults aged over 50:",conn.execute(s).fetchall())

del2 = ad_data.delete().where(ad_data.c.workclass != ' Private')
conn.execute(del2)
s = ad_data.select()
print()
print("Result after deleting records of adults not working in Private class:",conn.execute(s).fetchall())
```

4. Write two filter queries

```python
s = ad_data.select().where(ad_data.c.age > 40)
print("Fetching all adult details whose age is greater than 40:",conn.execute(s).fetchall())
print()
s2 = ad_data.select().where(ad_data.c.education == ' Masters')
print("Fetching all adult details whose education is 'Masters':",conn.execute(s2).fetchall())
```

5. Write two function queries

```python
from sqlalchemy.sql import func,select
print("Max age in all records is:",conn.execute(select([func.max(ad_data.c.age)])).fetchone())
print("Min age in all records is:",conn.execute(select([func.min(ad_data.c.age)])).fetchone())
```
